```python
import cv2
from random import randint
import numpy as np 
from glob import glob 
import time 
```

```python
image_folder = "/media/throgg/KINGSTON/Output/V1_20210622_082720/camera2/image_detail/"
output_folder = "/home/d/code/python/crack_detection/road_trainer/"
filenames = glob(image_folder+ "*.jpg")
width = 600
height = 400
# x = vertical 
xmin = 1500
xmax = 2500
# y = horizontal 
ymin = 1500
ymax = 2168

rgb = (200, 0, 0)
number = 0
for image in filenames:
    
    image= cv2.imread(image)
    
    scaled_image= cv2.rectangle(image, (xmin, ymax), (xmax, ymin), (0, 255, 0), 2)
    scaled_image = cv2.resize(image, (width, height))
    
    cropped_image = image[ymin:ymax, xmin:xmax]
    
    cv2.imshow("scaled", scaled_image)
    cv2.imshow("cropped", cropped_image)
    c = cv2.waitKey(0) % 256

    if c == ord('0'):
        cv2.imwrite(output_folder+str(randint(0, 10000))+".fine.jpg", cropped_image)
        print("saved fine image !")
    elif c == ord('1'):
        cv2.imwrite(output_folder+str(randint(0, 10000))+".broken.jpg", cropped_image)
        print("saved broken image !")
    else:
        print("image not saved")
        
```
