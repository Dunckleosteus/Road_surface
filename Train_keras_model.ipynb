```python
import numpy as np 
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
# ressource below 
# ressources for confusion matrix 
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
# https://www.youtube.com/watch?v=qFJeN9V1ZsI
```

```python
train_labels = [] 
train_samples = []
```

```python
for i in range (50):
    # the 5% of younder individuals with side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    # 5% of older participants with no side effects
    random_older = randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(0)
    
for i in range (1000):
    # 95% of younder with no side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    # 95% of older with side effects
    random_older = randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
# normalizing or standarizing data
scaler = MinMaxScaler(feature_range=(0,1))
# rescalling data to be between 0 and 1
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1, 1))
```

```python
for i in scaled_train_samples:
    print (i)
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')
print ("GPUs:", len(physical_devices))
```

```python
model= Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'),
    Dense(units=32, activation='relu'), # units is number of neurons
    Dense(units=2, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=["accuracy"])
```

```python
model.fit(x=scaled_train_samples, 
          y=train_labels, 
          batch_size=10, 
          epochs=30, 
          shuffle=True, # on by default 
          validation_split=0.1, # 10% of data put aside for validation
          verbose=2)
```

```python
# overfitting is when it's very good at recognising the
# models it has been trained on but not good with new dat
```

```python
predictions = model.predict(x=scaled_train_samples, batch_size=10, verbose=0)
```

```python
rounded_predictions = np.argmax(predictions, axis=-1)
```

```python
cm = confusion_matrix(
    y_true=train_labels,
    y_pred=rounded_predictions
)
cm
```
